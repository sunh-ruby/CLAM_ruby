```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def plot_auc_from_csvs(csv_dirs, group_names):
    if len(csv_dirs) != 3 or len(group_names) != 3:
        raise ValueError("Must provide exactly 3 CSV directories and 3 group names")
    
    plt.figure(figsize=(10, 6))
    
    for csv_dir, group_name in zip(csv_dirs, group_names):
        df = pd.read_csv(csv_dir)
        auc_values = df['test_auc']
        
        mean_auc = np.mean(auc_values)
        std_auc = np.std(auc_values)
        
        plt.bar(group_name, mean_auc, yerr=std_auc, capsize=5, alpha=0.7)
        plt.text(group_name, mean_auc, f'{mean_auc:.2f}Â±{std_auc:.2f}', 
                 ha='center', va='bottom')
    
    plt.ylabel('Mean AUC')
    plt.title('AUC Comparison With Different Amount of Data')
    plt.ylim(0, 1)  # AUC values are between 0 and 1
    plt.show()

# Example usage:
# csv_dirs = ['path/to/csv1.csv', 'path/to/csv2.csv', 'path/to/csv3.csv']
# group_names = ['Group A', 'Group B', 'Group C']
# plot_auc_from_csvs(csv_dirs, group_names)
```
